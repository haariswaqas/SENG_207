```python
# Importing all the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('data.csv')

# Remove null values
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Save the preprocessed data to a new CSV file
df.to_csv('preprocessed_data_file.csv', index=False)
```

```python
# Checking for null values and duplicates after pre-processing the data
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
# Filling null values with median
df['residual sugar'].fillna(df['residual sugar'].median(), inplace=True)
```

This code fills null values in the "residual sugar" column with the median value of that column. The median is a good measure of central tendency to use for filling null values because it is less affected by outliers than the mean.

By filling null values with the median value of the column, we can ensure that the imputed values are representative of the data in that column and do not significantly skew the distribution of that data. This can help to prevent bias in downstream analysis and modeling.

```python
# Obtain statistical inferences for each column
alcohol_mean = df['alcohol'].mean()
alcohol_std = df['alcohol'].std()

density_mean = df['density'].mean()
density_std = df['density'].std()

ph_mean = df['pH'].mean()
ph_std = df['pH'].std()

sulphates_mean = df['sulphates'].mean()
sulphates_std = df['sulphates'].std()

# Print the results
print(f'Alcohol column has a Mean of: {alcohol_mean}%')
print(f'Alcohol column has a Standard Deviation of: {alcohol_std}%')
print(" ")
print(f'Density column has a Mean of: {density_mean}')
print(f'Density column has a Standard Deviation of: {density_std}')
print(" ")
print(f'pH column has a Mean of: {ph_mean}')
print(f'pH column has a Standard Deviation of: {ph_std}')
```

Statistical Inferences

The alcohol column has a mean of 10.43% and a standard deviation of 1.067%.
The density column has a mean of 0.9967 and a standard deviation of 0.0019.
The pH column has a mean of 3.31 and a standard deviation of 0.155.
The sulphates column has a mean of 0.66 and a standard deviation of 0.17.
